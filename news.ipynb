```python
import pandas as pd
import numpy as np
import re
import string
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
def example_stopword_removal():
    sample_text = "The quick brown fox jumps over the lazy dog."
    words = sample_text.lower().split()
    filtered_words = [word for word in words if word not in stop_words]
    print("Original Text:", sample_text)
    print("After Stopword Removal:", " ".join(filtered_words))

example_stopword_removal()
```

```python
df_1 = pd.read_csv("True.csv")
df_1.dropna(inplace=True)
df_2 = pd.read_csv("Fake.csv")
df_2.dropna(inplace=True)
df = pd.concat([df_1, df_2], ignore_index=True)  # ignore_index=True resets the index
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(f"[{string.punctuation}]", "", text)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return " ".join(words)
```

```python
# Apply Preprocessing
df['processed_text'] = df['text'].apply(preprocess_text)
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    df['processed_text'], df['subject'], test_size=0.2, random_state=42)
```

```python
print(df.columns)
```

```python
# Vectorization
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
# Model Training
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)
```

```python
# Prediction
y_pred = model.predict(X_test_tfidf)
```

```python
# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
